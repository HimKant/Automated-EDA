## Create a function Graph that takes 4 arguments - data,column,directory ,a number to differentiate discrete columns from continuous columns and does bivariate analysis.

```python
def Graph(data1,n,columns='', directory = os.getcwd()):
    
    if columns == '': 
        #if no columns assigned then all columns should be selected
        data = data1
    else:
        #else selected columns should be there in dataframe
        data = data1[columns]
        
    #counting missing values 
    ## null count analysis
    import missingno as msno
    msno.bar(data1,color='teal')
    plt.xlabel('Columns',fontsize=15)
    plt.ylabel("Count",fontsize=15)
    plt.title("Missing values",fontsize=20)
    plt.savefig(directory + r'/missing_values.png',dpi=1000,orientation ='landscape',pad_inches = 1,facecolor='yellow') 
    plt.show()
    
    
    categorical = []                                 

    # checking whether the variables are numerical or categorical 
    for col in data.columns:                  
        if data.dtypes[col] == 'object':
            categorical.append(col)                

    # plotting bar chart for the categorical variables
    for j in categorical:
        data['%s' %j].value_counts().plot(kind = 'bar', figsize=(10,6),color="indigo")
        plt.ylabel('count')
        plt.title("{} Bar Chart".format(j),fontsize=20)
        plt.savefig(directory + r'/%s.png' %j,dpi=1000,orientation ='landscape',pad_inches = 1,facecolor='yellow') 
        plt.show()   
        
        
    numerical = []
    discrete = []
    
    for column in data.columns:
        if data[column].dtype in (int,float) and len(data[column].unique()) > n:
            numerical.append(column)
        else:
            discrete.append(column)
     
    #plotting bar plot for discerete variables
    for k in discrete:
        data['%s' %k].value_counts().plot(kind = 'bar', figsize=(10,6),color="coral")
        plt.ylabel('count',fontsize=15)
        plt.xlabel('%s'%k,fontsize=15)
        plt.title("{} Bar Chart".format(k),fontsize=20)
        plt.savefig(directory + r'/%s.png' %k,dpi=1000,orientation ='landscape',pad_inches = 1,facecolor='yellow') 
        plt.show()
        
    # plotting histogram and boxplot for each numerical variable
    for i in numerical:
        plt.subplot(1,1,1) #(row, column, plot)
        # plotting histogram for each numerical variable
        data['%s'%i].plot(kind='hist',alpha=0.5, figsize=(8,6), color="Green",grid=True,edgecolor='black',density = True)
        data['%s'%i].plot(kind='kde',linewidth=2.5,color="black")   
        plt.xlabel('%s'%i,fontsize=15)
        plt.ylabel("Frequency",fontsize=15)
        plt.title("{} Histogram".format(i),fontsize=20)
        plt.savefig(directory + r'/%s.png' %i,dpi=1000,orientation ='landscape',pad_inches = 1,facecolor='yellow') 
        plt.show()

        plt.subplot(2,1,1)
        # plotting boxplot for each numerical variable
        data.boxplot(i,figsize=(8,6),notch=True,grid=True,vert=False,fontsize=15,color="Green")
        plt.title("{} Boxplot".format(i),fontsize=20)
        plt.savefig(directory + r'/%s.png' %i,dpi=1000,orientation ='landscape',pad_inches = 1,facecolor='yellow') 
        plt.show()

    #Relationship between two categorical variables
    for i in range(0,len(categorical)):
        pri=categorical[i]
        for j in range(0,len(categorical)):
            sec=categorical[j]
            if sec!=pri:
                plt.figure(figsize=(8,6))
                plot=sns.countplot(data=data,x=pri,hue=sec,palette='Accent')


    #Relationship between categorical and numerical variables
    for i in range(0,len(categorical)):
        cat=categorical[i]
        for j in range(0,len(numerical)):
            num=numerical[j]
            plt.figure(figsize=(8,6))
            sns.boxplot(x=cat,y=num,data=data,palette="flare")
            
    #Relationship between numerical and numerical variables
    for i in range(0,len(numerical)):
        num1=numerical[i]
        for j in range(0,len(numerical)):
            num2=numerical[j]
            if num1!=num2:
                plt.figure(figsize=(8,6))
                sns.color_palette("rocket", as_cmap=True)
                sns.scatterplot(x=num1,y=num2,data=data)
                
    #using correlation matrix to find correlation of all numeric columns.
    correlation=data.corr()
    sns.heatmap(correlation,cmap="PiYG",annot=True)
    plt.savefig(directory + r'/heatmap.png',dpi=1000,pad_inches = 1) 
    plt.show()

    # Pair plot or scatterplot is a good complementary to the 
    #correlation matrix, especially when nonlinear relationships 
    #(e.g. exponential, inverse relationship) might exist.
    sns.pairplot(data)
    plt.savefig(directory + r'/pairplot.png',dpi=1000,pad_inches = 1) 
    plt.show()
```
